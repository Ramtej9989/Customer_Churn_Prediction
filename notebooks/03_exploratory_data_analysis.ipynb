```python
df = pd.read_csv("../data/processed/customer_churn_cleaned.csv")
```

```python
df["Churn"].value_counts()
```

```python
sns.countplot(data=df, x="Churn")
plt.title("Customer Churn Distribution")
plt.show()
```

```python
(df["Churn"].value_counts(normalize=True) * 100).round(2)
```

## Observation

- Most customers did not churn.
- Around 46.74% of customers churned.

## Observation

- Customers with Month-to-Month contracts show the highest churn.
- Customers with One-Year and Two-Year contracts are more likely to stay.
- Long-term contracts improve customer retention.

# Key Business Insights

1. Customers with Month-to-Month contracts have the highest churn rate.

2. Customers paying higher monthly charges are more likely to churn.

3. Customers with shorter tenure are at greater risk of churning.

4. Lower satisfaction scores are strongly associated with churn.

5. Customers who raise more support tickets tend to churn more often.

6. Gender has little impact on customer churn.

7. Long-term contracts help improve customer retention.

8. Fiber Optic customers show a higher churn rate compared to other internet service types.
