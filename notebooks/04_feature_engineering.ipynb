```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/customer_churn_cleaned.csv")

df.head()
```

```python
bins = [0, 12, 24, 48, 72]
labels = ["0-1 Year", "1-2 Years", "2-4 Years", "4-6 Years"]

df["TenureGroup"] = pd.cut(
    df["TenureMonths"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
df["TenureGroup"].value_counts()
```

```python
df["MonthlyChargeCategory"] = pd.qcut(
    df["MonthlyCharges"],
    q=3,
    labels=["Low", "Medium", "High"]
)
```

```python
df["MonthlyChargeCategory"].value_counts()
```

```python
df["SupportLevel"] = pd.cut(
    df["SupportTickets"],
    bins=[-1, 1, 3, 10],
    labels=["Low", "Medium", "High"]
)
```

```python
df["SupportLevel"].value_counts()
```

```python
df["AverageMonthlySpend"] = (
    df["TotalCharges"] / df["TenureMonths"]
)
```

```python
df["AverageMonthlySpend"] = (
    df["AverageMonthlySpend"]
    .replace([np.inf, -np.inf], 0)
    .fillna(0)
)
```

```python
df.to_csv(
    "../data/processed/customer_churn_feature_engineered.csv",
    index=False
)
```

# Feature Engineering Summary

## New Features Created

These features capture customer behavior in a more meaningful way than the raw values alone. They may improve model performance and make the model's decisions easier to interpret.
