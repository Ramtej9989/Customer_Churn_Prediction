```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/processed/customer_churn_feature_engineered.csv")

df.head()
```

```python
df.info()
```

```python
df["Churn"] = df["Churn"].map({
    "No":0,
    "Yes":1
})
```

```python
df["Churn"].value_counts()
```

```python
df.drop("CustomerID", axis=1, inplace=True)
```

```python
X = df.drop("Churn", axis=1)

y = df["Churn"]
```

```python
categorical_cols = X.select_dtypes(include="object").columns

categorical_cols
```

```python
categorical_cols = X.select_dtypes(include=["object", "category"]).columns

categorical_cols
```

```python
X = pd.get_dummies(
    X,
    columns=categorical_cols,
    drop_first=True
)
```

```python
df.head
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("Training Features:", X_train.shape)
print("Testing Features :", X_test.shape)
print("Training Target  :", y_train.shape)
print("Testing Target   :", y_test.shape)
```

```python
numerical_cols = X_train.select_dtypes(include=["int64", "float64"]).columns
scaler = StandardScaler()
```

```python
X_train[numerical_cols] = scaler.fit_transform(
    X_train[numerical_cols]
)

X_test[numerical_cols] = scaler.transform(
    X_test[numerical_cols]
)
```

```python
X_train.head()
```

```python
X_train.info()
```

```python
X_train.to_csv("../data/processed/X_train.csv", index=False)
X_test.to_csv("../data/processed/X_test.csv", index=False)

y_train.to_csv("../data/processed/y_train.csv", index=False)
y_test.to_csv("../data/processed/y_test.csv", index=False)
```

# Data Preprocessing Summary

## Steps Performed

- Converted the target variable (`Churn`) into numeric values.
- Removed the `CustomerID` column.
- Separated features and target.
- Applied One-Hot Encoding to categorical variables.
- Split the dataset into training and testing sets using an 80:20 ratio.
- Standardized numerical features using `StandardScaler`.
- Saved the processed datasets for model training.

## Result

The dataset is fully prepared for machine learning.
