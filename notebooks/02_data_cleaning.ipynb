```python
import pandas as pd

df = pd.read_csv("../data/raw/customer_churn.csv")
df.head
```

```python
df.isnull().sum()
```

```python
df[df["InternetService"].isnull()]
```

```python
df["InternetService"].value_counts(dropna=False)
```

```python
df["InternetService"] = df["InternetService"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/processed/customer_churn_cleaned.csv", index=False)
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
numerical_cols
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

# Data Cleaning Summary

- Loaded the raw dataset.
- Identified missing values.
- Replaced missing values in `InternetService` with `Unknown`.
- Verified data types.
- Confirmed there were no duplicate records.
- Inspected numerical features for outliers using boxplots.
- Chose not to remove outliers because they represent valid customer behavior.
- Saved the cleaned dataset for further analysis.
